## Search for people in the test sample that have companions in the train sample

```python
df['PassengerId'].count()
```

```python
dfTest['PassengerId'].count()
```

```python
arr_tickets=[]
for i in range(418):       
    ticket1 = dfTest.loc[i]['Ticket']
    for j in range(891):
        ticket2 = df.loc[j]['Ticket']
        if (ticket1==ticket2):
            arr_tickets.append(ticket1)
            continue
```

```python
arr_tickets_set=set(arr_tickets)
arr_tickets_list=list(arr_tickets_set)
```

```python
np_arr = np.array(arr_tickets_list)
np_arr.size
```

#### Find people in the test set that have survived campanions in the test set

```python
arr_ticks_surv=[]
for k in range(np_arr.size):
    cond1=df['Ticket']==arr_tickets_list[k]
    cond2=df['Survived']>0
    if (df[(cond1)&(cond2)]['PassengerId'].count()>0): arr_ticks_surv.append(arr_tickets_list[k])
```

```python
np.array(arr_ticks_surv).size
```

#### Check manually

```python
for i in range(69):
    print(' ')
    print('group # {}'.format(i))
    print(dfTest[dfTest['Ticket']==arr_ticks_surv[i]][['PassengerId','Pclass','Sex','Age','Ticket','Parch','SibSp']])
    print(df[df['Ticket']==arr_ticks_surv[i]][['PassengerId','Pclass','Sex','Age','Ticket','Survived','Parch','SibSp']])
```

#### Now check other tickets

```python
arr_ticks_dead=[]
for k in range(np.array(arr_tickets_list).size):
    found=0
    for l in range(np.array(arr_ticks_surv).size):
        if (arr_tickets_list[k]==arr_ticks_surv[l]): 
            found=1
            break
    if (found==0): arr_ticks_dead.append(arr_tickets_list[k])
    
np.array(arr_ticks_dead).size
```

```python
for i in range(46):
    print(' ')
    print('group # {}'.format(i))
    print(dfTest[dfTest['Ticket']==arr_ticks_dead[i]][['PassengerId','Pclass','Sex','Age','Ticket','Parch','SibSp','Embarked']])
    print(df[df['Ticket']==arr_ticks_dead[i]][['PassengerId','Pclass','Sex','Age','Ticket','Survived','Parch','SibSp','Embarked']])
```

## 6. Prepare the submition file

```python
dfNew=pd.DataFrame(columns=('PassengerId', 'Survived'))

survivedId=[941,1009,932,225,924,1246,896,161,283,1046,1271,165,127,230,996,931,333]

for i in range(418):
    surv=0       
    PassId=dfTest.loc[i]['PassengerId']
    if (dfTest.loc[i]['Sex']=='female'): 
        if (dfTest.loc[i]['Pclass']==1): surv=1
        if (dfTest.loc[i]['Pclass']==2): surv=1
        if (dfTest.loc[i]['Pclass']==3): 
            if (dfTest.loc[i]['Embarked']=='Q'): surv=1
                
    if (dfTest.loc[i]['Sex']=='male'):
        if (dfTest.loc[i]['Pclass']==1): 
            if (dfTest.loc[i]['Age']<14): surv=1
        if (dfTest.loc[i]['Pclass']==2):
            if (dfTest.loc[i]['Age']<14): surv=1
                
    if(dfTest.loc[i]['PassengerId'] in survivedId): surv=1
            
    dfNew.loc[i] = pd.Series({'PassengerId':PassId,'Survived':surv})

dfNew.to_csv('submit.csv',index=False)
# 0.77033
```
