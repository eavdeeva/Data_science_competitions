```python
def SepMeanStd(m1,m2,s1,s2):
    if ((s1+s2)==0): return 0
    return (abs(m1-m2))/math.sqrt(s1*s1+s2*s2)

def CatValueSeparations(df,featureName):
    vals=df[featureName].unique()
    means=np.zeros(vals.size)
    stds=np.zeros(vals.size)
    seps=np.zeros(vals.size)
    for j in range(vals.size):
        dfLocal = df[df[featureName]==vals[j]]
        means[j]=dfLocal['SalePrice'].mean(axis=0)
        stds[j]=dfLocal['SalePrice'].std(axis=0)
        if(stds[j]!=stds[j]): stds[j]=0
    for k in range(vals.size):
        for i in range(vals.size):
            if (means[i]<means[k]):
                buff=means[i]
                means[i]=means[k]
                means[k]=buff
                buff=stds[i]
                stds[i]=stds[k]
                stds[k]=buff
                buff=vals[i]
                vals[i]=vals[k]
                vals[k]=buff
    print('Sorted by mean:')
    for j in range(vals.size):
        n = df[df[featureName]==vals[j]].shape[0]
        print('{}, n={}: {} +- {}'.format(vals[j],n,int(means[j]),int(stds[j])))
        if (j+1<vals.size):
            seps[j]=SepMeanStd(means[j],means[j+1],stds[j],stds[j+1])
            print(seps[j])
        else:
            seps[j]=SepMeanStd(means[j],means[0],stds[j],stds[0])
            print(seps[j])
    return (vals,seps)
```
