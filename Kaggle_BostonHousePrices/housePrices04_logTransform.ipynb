```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv('all/train.csv')
```

```python
from scipy.stats import norm

#list_num_feat=['SalePrice','MiscVal','PoolArea','LotArea',
#               'LowQualFinSF','3SsnPorch','KitchenAbvGr',
#              'BsmtFinSF2',ScreenPorch']
list_num_feat=['SalePrice','MiscVal','PoolArea','LotArea',
              'LowQualFinSF','3SsnPorch','KitchenAbvGr',
              'BsmtFinSF2','ScreenPrch']

for i in range(len(list_num_feat)):
    fig, (ax1,ax2) = plt.subplots(figsize=(10,3),ncols=2)
    s = list_num_feat[i]
    s_l = list_num_feat[i]+'_Log'
    df[s_l] = df[s].apply(lambda x: np.log1p(x))
    sns.distplot(df[s],fit=norm,ax=ax1)
    sns.distplot(df[s_l],fit=norm,ax=ax2)
    
```
