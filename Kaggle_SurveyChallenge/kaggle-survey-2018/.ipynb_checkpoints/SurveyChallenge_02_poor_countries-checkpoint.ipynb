```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv('multipleChoiceResponses.csv')
df.drop([0],inplace=True)
df.head()
```

```python
file=open('multipleChoiceResponses.csv','r')
l1 = file.readline()
l2 = file.readline()
l1_cols = l1.split(',')
l2_cols = l2.split(',')
for col in range(len(l1_cols)):
    print (l1_cols[col],l2_cols[col])
```

```python
df['Time'] = df['Time from Start to Finish (seconds)'].apply(int)
df_time = df['Time']
df_time.head()
time_min = df_time.unique().min()
time_max = df_time.unique().max()
time_mean = df_time.unique().mean()
time_std = df_time.unique().std()
print(time_min)
print(time_max)
print(int(time_mean))
print(int(time_std))


ax1 = plt.xlim(0,40000)
ax1 = plt.ylim(0,2000)
ax1 = plt.hist(df_time,bins=500)
```

```python
df_fast = df[df['Time']<1800]['Time']

ax2 = plt.hist(df_fast,bins=100)
```
